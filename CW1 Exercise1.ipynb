```python
import matplotlib.pyplot as plt
# Function to convert dates to months between 0-648
# which is stored in dictionary ‘x’ and then months are plotted against prices.
mnts = []
for date in range(0,len(nfdates)):
    mnts.append(date)
plt.plot(mnts,price)
plt.title('Djia price between the months of 1914-1968')
plt.xlabel('Months')
plt.ylabel('Price')
plt.grid(True)
    
```

```python

import matplotlib.pyplot as plt
#creates an x,y array of months and price, allowing them to be plotted directly
x = np.array(mnts)
y = np.array(price)


#Calculating coefficients
coeff = np.polyfit(x,y,1)

#Equation of the line for line of best fit 
#y_line = m*xv+c

#creates scatter plot 
plt.figure(figsize=(10,6))
plt.scatter(x,y, label='DJIA data')

#poly1d creates a polynomial from equation of line coefficients
#Stored in variable djiapricep(stands for djiapriceprediction)
best_fit=np.poly1d(coeff)

#Plot line of best fit 
plt.plot(x,best_fit(x), color='black', label='Line of best fit')

#Display equation of the line on the graph 
#plt.text(1,800, 'y = ' + '{:.2f}'.format(m) + ' + {:.2f}'.format(c) + 'x', size=13)

#Add axis labels and titles 
plt.title('Line of best fit DJIA data')
plt.xlabel('Months')
plt.ylabel('Price')
plt.legend()
plt.grid(True)



```
